```python
#from sklearn import linear_model
from sklearn.linear_model import LinearRegression

# Datos dados
x = np.linspace(1, 19, 10).reshape(-1, 1)
y = np.linspace(17, 53, 10)

# Crear un modelo de regresión lineal
regression_model = LinearRegression()

# Entrenar el modelo de regresión lineal
regression_model.fit(x, y)

# Imprimir los coeficientes de la regresión
print("Coeficiente (pendiente):", regression_model.coef_[0])
print("Término independiente:", regression_model.intercept_)
```

Por lo tanto, en el planeta X la relación es:

**F = 2m + 15**

```python
pred = regression_model.predict(nuevos)
pred
```

## 2 datos de entrada y 1 target (caso 3D)

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generar datos aleatorios con tendencia lineal en 3D
np.random.seed(0)
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = np.random.rand(n_samples)
y = 2 * x1 + 3 * x2 + -5 #np.random.rand(n_samples)  # tendencia lineal en 3D

# Convertir los datos en una matriz de características (2D) y un vector de destino (1D)
X = np.column_stack((x1, x2))
X
```

```python
# Crear un modelo de regresión lineal
model = LinearRegression()

# Ajustar el modelo a los datos
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# y = 2 * x1 + 3 * x2 + -5
prueba = np.array([(1,0),(0,1),(0,0),(1,1)])
resultado = model.predict(prueba)
resultado
```

```python
# Hacer predicciones para nuevas entradas
x1_new = np.random.rand(10)
x2_new = np.random.rand(10)
X_new = np.column_stack((x1_new, x2_new))
y_pred = model.predict(X_new)

# Graficar los datos y la superficie de regresión
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar los datos originales
ax.scatter(x1, x2, y, c='blue', marker='o')

# Graficar las predicciones
ax.scatter(x1_new, x2_new, y_pred, c='red', marker='x')

# Configurar las etiquetas de los ejes
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.title('Datos originales y predicciones')
plt.legend(['Datos originales', 'Predicciones'])
plt.show()
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# Semilla para reproducibilidad
np.random.seed(0)

# Datos
n_samples = 100
x1 = np.random.rand(n_samples)
x2 = np.random.rand(n_samples)
y = 2 * x1 + 3 * x2 - 5

# Convertir los datos en una matriz de características (2D) y un vector de destino (1D)
X = np.column_stack((x1, x2))

# Modelo de red neuronal
model = Sequential()
model.add(Dense(2, input_dim=2, activation='linear'))
model.add(Dense(12, activation='linear'))
model.add(Dense(24, activation='linear'))
model.add(Dense(12, activation='linear'))
model.add(Dense(1, activation='relu'))

# Compilar el modelo
model.compile(optimizer='adam',
              loss = 'mean_squared_error',
              metrics = 'accuracy')

# Entrenar el modelo
historial = model.fit(X, y, epochs=200, batch_size=10, verbose=0)

# Hacer predicciones
predictions = model.predict(X)
```

```python
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"]);
```
